## Segment 1 - Defining elements of a plot

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
from matplotlib import rcParams
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
rcParams['figure.figsize'] = (5,4)
```

### Define axes, ticks, and grids

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0, 4, 3, 2, 1]

fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.set_xlim([1, 9])
ax.set_ylim([0, 5])

ax.set_xticks([0, 1, 2, 4, 5, 6, 8, 9, 10])
ax.set_yticks([0, 1, 2, 3, 4, 5])

ax.plot(x, y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1, .1, 1, 1])
ax.set_xlim([1, 9])
ax.set_ylim([0, 5])

ax.grid()
ax.plot(x, y)
```

### Generating multiple plots in one figure

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(x)
ax2.plot(x, y)
```
