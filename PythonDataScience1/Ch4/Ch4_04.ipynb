## Segment 4 - Creating labels and annotations

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
from matplotlib import rcParams
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
rcParams['figure.figsize'] = (8,4)
```

### Labeling plot features

#### The functional method

```python
x = range(1, 10)
y = [1, 2, 3, 4, 0.5, 4, 3, 2, 1]
plt.bar(x, y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1, 2, 3, 4, 0.5]
veh_type = ['bicycle', 'motorbike', 'car', 'van', 'stroller']

plt.pie(z, labels=veh_type)
plt.show()
```

#### The object-oriented method

```python
address = '/Users/mattjwilliams/Documents/LinkedInDataScience/PythonDataScience1/PFDS1/Data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
```

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per gallon of cars in mtcars.csv')
ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')
ax.grid()
```

### Adding a legend

#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc='best')
plt.show()
```

#### The object-oriented method

```python
fig = plt.figure()
ax = fig.add_axes([0.1, 0.1, 0.9, 0.9])
mpg.plot()
ax.set_xticks(range(32))
ax.set_xticklabels(cars.car_names, rotation=60, fontsize='medium')
ax.set_title('Miles per gallon of cars in mtcars.csv')
ax.set_xlabel('car names')
ax.set_ylabel('miles per gallon')
ax.grid()
ax.legend(loc='best')
```
