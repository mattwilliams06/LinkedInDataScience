## Segment 6 - Creating statistical data graphics

```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame
from matplotlib import rcParams
```

```python
%matplotlib inline
%config InlineBackend.figure_format = 'retina'
rcParams['figure.figsize'] = (5,4)
```

```python
import seaborn as sb
sb.set_style('whitegrid')
```

```python
address = '/Users/mattjwilliams/Documents/LinkedInDataScience/PythonDataScience1/PFDS1/Data/mtcars.csv'
cars = pd.read_csv(address)
cars.columns = ['car_names', 'mpg', 'cyl', 'disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']
cars.index = cars.car_names
mpg = cars['mpg']
mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

### Scatterplots

```python
cars.plot(kind='scatter', x='hp', y='mpg', c=['darkgray'], s=150)
```

```python
sb.regplot(x='hp', y='mpg', data=cars, scatter=True)
```

### Generating a scatterplot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Building a boxplot

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
