## Segment 6 - Delving into non-parametric methods with pandas and scipy

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
from pylab import rcParams

import scipy
from scipy.stats import spearmanr
```

### The Spearman Rank Correlation

```python
address = '/Users/mattjwilliams/Documents/LinkedInDataScience/PythonDataScience1/PFDS1/Data/mtcars.csv'

cars = pd.read_csv(address)
cars.columns = ["car_names","mpg","cyl","disp","hp","drat","wt","qsec","vs","am","gear","carb"]
```

```python
cars.head()
```

```python
sb.pairplot(cars)
```

```python
X = cars[['cyl', 'vs', 'am', 'gear']]
sb.pairplot(X)
```

```python
cyl = cars['cyl']
vs = cars['vs']
am = cars['am']
gear = cars['gear']

spearmanr_coefficient, p_value = spearmanr(cyl, vs)

print(f'SpearmanR coefficient: {spearmanr_coefficient:.3f}')
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, am)

print(f'SpearmanR coefficient: {spearmanr_coefficient:.3f}')
```

```python
spearmanr_coefficient, p_value = spearmanr(cyl, gear)

print(f'SpearmanR coefficient: {spearmanr_coefficient:.3f}')
```

### Chi-square test for independence

```python
table = pd.crosstab(cyl, am)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic: {chi2:.3f}, p-value: {p:.3f}')
```

```python
table = pd.crosstab(cyl, vs)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic: {chi2:.3f}, p-value: {p:.3f}')
```

```python
table = pd.crosstab(cyl, gear)
from scipy.stats import chi2_contingency

chi2, p, dof, expected = chi2_contingency(table.values)
print(f'Chi-square statistic: {chi2:.3f}, p-value: {p:.3f}')
```
