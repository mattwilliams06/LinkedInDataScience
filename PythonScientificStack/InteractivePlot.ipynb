```python
from ipywidgets import interact
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
@interact(limit=6)
def plot_sin(limit):
    xs = np.linspace(-limit, limit, 100)
    ys = np.sin(xs)
    plt.plot(xs, ys, label=f'sin(x), {-limit} - {limit}')
    plt.show()
```
