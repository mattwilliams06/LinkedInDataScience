```python
import pyarrow.parquet as pq
```

```python
address = '/Users/mattwilliams/Documents/LinkedInLearning/TipsTricksTech/Exercise Files/data/MonthlyProductSales.parquet'
```

```python
table = pq.read_table(address)
```

```python
table_dict = dict(table.to_pydict())
```

```python
for k, v in table_dict.items():
    print(k)
```

```python
keys = [k for k in table_dict.keys()]
keys
```

```python
values = [v for v in table_dict.values()]
```

```python
pivoted_values = zip(*values)
```

```python
table_dictionary_array = []
for record in pivoted_values:
    table_dictionary_array.append(dict(zip(keys, record)))
```

```python
for record in table_dictionary_array:
    print(record)
```
